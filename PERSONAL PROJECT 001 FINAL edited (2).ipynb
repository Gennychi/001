<h3>CONTRIBUTING FACTORS THAT AFFECTS RANKING</h3>
Let's explore the metrics used to gauge these universities!

The six metric used in ranking these universities across the world include:

*Academic Reputation (40%)
*Employer Reputation (10%)
*Faculty/Student Ratio (20%)
*Citations per faculty (20%)
*International Faculty Ratio (5%)
*International Student Ratio (5%)

Out of the 6 metrics that have been used, only 3 are present in this dataset.
Research output - 20%
Student Faculty ratio - 20%
International students - 5%
So, our analysis will account only for 45% of the survey methodology.


<h3>Research output per region</h3>

```python
#cleaning the research_output column and a creating a boolean for the university type column
new_df = university_df.copy()
new_df['research_output'] = new_df['research_output'].replace(['Very high'], 'Very High')
new_df["yes public"] = new_df["type"]
new_df["yes public"]=new_df["yes public"].replace(["Public"], "yes")
new_df["yes public"]=new_df["yes public"].replace(["Private"], "no")
new_df.head()
```

Public universities produces very high research output more than Private univesrities
