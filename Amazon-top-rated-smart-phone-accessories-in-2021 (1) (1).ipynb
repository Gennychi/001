<ul>

**RENAMING COLUMNS**

```python
# Renaming the Columns in Dataset in a suitable Pythonic way
amazon_df= amazon_df.rename(columns = {'Smartphone': 'PhoneModel', 'Reviews count': 'ReviewsCount',
                                       'Offer price': 'OfferPrice', 'Actual price': 'ActualPrice',
                                       'Status of stock & more offers': 'StatusOfStock'},inplace= False)
```

```python
amazon_df.head(50)
```

<h2>Fix and format the ratings,savings, reviews count, offer price, actual price columns</h2>

```python
#Further Data cleaning. Here we will remove the commas & currency signs in the ReviewsCount, OfferPrice and ActualPrice column.
amazon_df['ReviewsCount'] = amazon_df['ReviewsCount'].str.replace(",","")
amazon_df['OfferPrice'] = amazon_df['OfferPrice'].str.replace(",","")
amazon_df['OfferPrice'] = amazon_df['OfferPrice'].str.replace("₹","")
amazon_df['ActualPrice'] = amazon_df['ActualPrice'].str.replace(",","")
amazon_df['ActualPrice'] = amazon_df['ActualPrice'].str.replace("₹","")
amazon_df.head()
```

**<h3>Filling the Mising values</h3>**

```python
new_amazondf = amazon_df.copy()
new_amazondf['Delivery']=new_amazondf['Delivery'].fillna(new_amazondf['Delivery'].mode()[0])
new_amazondf['StatusOfStock']=new_amazondf['StatusOfStock'].fillna(new_amazondf['StatusOfStock'].mode()[0])
new_amazondf=new_amazondf.fillna(0)
new_amazondf.tail(12)
```

**<h3>Convert the strings columns to Numerical columns</h3>**

Convert`ReviewsCount` ,`ActualPrice`,`OfferPrice` Columns to numerical columns

```python
new_amazondf.info()
```

```python
new_amazondf['ReviewsCount'] = new_amazondf['ReviewsCount'].astype('float').astype('int')
new_amazondf['OfferPrice'] = new_amazondf['OfferPrice'].astype('float').astype('int')
new_amazondf['ActualPrice'] = new_amazondf['ActualPrice'].astype('float').astype('int')

```

```python
#This is the new data types after converting some columns from strings to integers
new_amazondf.info()
```

<h3> Creating a new column for the % Saved</h3>

```python
#Creating a new column for % saved 

new_amazondf['savings']= (new_amazondf['ActualPrice']- new_amazondf['OfferPrice'])/ new_amazondf['ActualPrice']
new_amazondf['savingsby%'] = (new_amazondf['savings'] *100)
new_amazondf.head().round(0)
```

```python
#Dropping the columns i dont need
new_df= new_amazondf.drop(['Saving','savings'], axis = 1)
new_df.head().round(0)
```
